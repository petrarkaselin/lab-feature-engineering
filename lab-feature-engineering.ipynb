# Activity

Create a function that takes a dataframe as an input and a percentual threshold (default value = 0.25) and returns a list of columns with null values greater than the specified threshold

````python
def columns_with_high_null_values(df, threshold=0.25):

    total_rows = len(df)
    null_threshold = int(threshold * total_rows)
    high_null_columns = df.columns[df.isnull().sum() > null_threshold].tolist()
    
    return high_null_columns
````

## Replacing NaNs in 'WEALTH1', 'ADI', 'DMA', 'MSA' using KNNImputer

````python
from sklearn.impute import KNNImputer

impute = ['WEALTH1', 'ADI', 'DMA', 'MSA']

imputation_df = data[impute].copy()

# Initialize KNNImputer
for n_neighbors in range(5, 21, 5):
    imputer = KNNImputer(n_neighbors=n_neighbors)
    imputation_df[impute] = imputer.fit_transform(imputation_df[impute])
````
